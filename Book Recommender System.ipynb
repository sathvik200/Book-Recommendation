```python
x = ratings['user_id'].value_counts()>200
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on ='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['ratings'].count().reset_index()
```

```python
final_rating = ratings_with_books.merge(number_rating, on = 'title')
```

```python
final_rating.rename(columns={'ratings_y': 'number_of_ratings'}, inplace=True)
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_names= final_rating['title'].unique()
```

```python
book_names
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='ratings_x')
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1), n_neighbors=6)
```

```python
  
suggestions.tolist()
for i in suggestions:
    print(book_pivot.index[i])
```

```python
recommended_urls = []
recommended_books = []
for i in range(len(suggestions)):
    for j in books['title']:
        if (j == book_pivot.index[suggestions[i]]).any():
            x = books[books['title'] == j]['Image-URL-M'].reset_index()
            x.drop(columns = 'index', inplace = True)
            if x['Image-URL-M'][0] not in recommended_urls:
                recommended_urls.append(x['Image-URL-M'][0])    
        recommended_books.append(book_pivot.index[suggestions[i]])
```
